```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("Unemployment in India.csv")

#  FIX column name issue
df.columns = df.columns.str.strip()

# Check columns
print(df.columns)

# Convert Date column
df['Date'] = pd.to_datetime(df['Date'])

# Drop missing values
df.dropna(inplace=True)

# Plot unemployment trend
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Estimated Unemployment Rate (%)'])
plt.title("Unemployment Rate in India")
plt.xlabel("Date")
plt.ylabel("Unemployment Rate (%)")
plt.show()
```

```python
covid_start = "2020-03-01"

pre_covid = df[df['Date'] < covid_start]
during_covid = df[df['Date'] >= covid_start]

print("Pre-Covid Avg Unemployment:",
      round(pre_covid['Estimated Unemployment Rate (%)'].mean(), 2))
print("During-Covid Avg Unemployment:",
      round(during_covid['Estimated Unemployment Rate (%)'].mean(), 2))
```

```python
from sklearn.linear_model import LinearRegression

# Convert Date
df['date_ordinal'] = df['Date'].map(pd.Timestamp.toordinal)

X = df[['date_ordinal']]
y = df['Estimated Unemployment Rate (%)']

model = LinearRegression()
model.fit(X, y)

future_dates = pd.date_range(df['Date'].max(), periods=6, freq='M')
future_ordinals = future_dates.map(pd.Timestamp.toordinal)

future_predictions = model.predict(future_ordinals.values.reshape(-1, 1))

print("Predicted unemployment rates:", future_predictions)
```
